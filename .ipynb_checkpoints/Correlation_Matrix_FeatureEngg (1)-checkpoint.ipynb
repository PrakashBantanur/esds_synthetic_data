```python

uri = ["login.php", "index.php", "vulnerabilities", "exec", "dvwa", "js", "add_event_listeners.js", "csrf", "upload", 
           "captcha", "sqli", "sqli_blind", "fi", "?page=include.php", "brute", "weak_id", "xss_d", "xss_r", "xss_s"]

uri2 = ["/", "/login.php", "/index.php", "/vulnerabilities/exec/", "/dvwa/js/add_event_listeners.js", 
        "/vulnerabilities/csrf/", "/vulnerabilities/upload/", "/vulnerabilities/captcha/", "/vulnerabilities/sqli/",
        "/vulnerabilities/sqli_blind/", "/vulnerabilities/fi/?page=include.php", "/vulnerabilities/brute/",
        "/vulnerabilities/weak_id/", "/vulnerabilities/xss_d/", "/vulnerabilities/xss_r/", "/vulnerabilities/xss_s/"]

print(len(uri))
print(len(uri2))

for i in range(len(df)):
    key1 = df.loc[i, 'requested_uri']

    l=[]
    m=[]
    for p in range(len(uri)):
        dist = leven.distance(str(key1), str(uri[p]))
        l.append(dist)
    uri1average = round((sum(l) / len(l)), 3)
#     print(l)
#     print(uri1average)

    for q in range(len(uri2)):
        dist2 = leven.distance(str(key1), str(uri2[q]))
        m.append(dist2)
    uri2average = round((sum(m) / len(m)), 3)
#     print(m)
#     print(uri2average)
    
    df.loc[i, 'Avg_LevenDist1'] = uri1average
    df.loc[i, 'Avg_LevenDist2'] = uri2average
    
df
```
